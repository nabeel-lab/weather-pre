```python
year=pd.to_datetime(raw_df['Date']).dt.year
train_df=raw_df[year<2015]
val_df=raw_df[year==2015]
test_df=raw_df[year>=2015]
```

```python
from sklearn.model_selection import train_test_split
train_val_df,test_df=train_test_split(test_df,test_size=0.2,random_state=42)
train_df,val_df=train_test_split(train_val_df,test_size=0.25,random_state=42)
```

```python
input_cols=train_df.columns[1:-1]
target_col=train_df.columns[-1]
input_cols,target_col
```

```python
train_inputs=train_df[input_cols].copy()
train_targets=train_df[target_col].copy()
val_inputs=val_df[input_cols].copy()
val_targets=val_df[target_col].copy()
test_inputs=test_df[input_cols].copy()
test_targets=test_df[target_col].copy()
```

```python
train_inputs.head()
```

```python
numarical_cols=train_inputs.select_dtypes(include=np.number).columns.tolist()
categorical_cols=train_inputs.select_dtypes('object').columns.tolist()
categorical_cols
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

imputer=SimpleImputer(strategy='median')
imputer.fit(train_inputs[numarical_cols])
train_inputs[numarical_cols]=imputer.transform(train_inputs[numarical_cols])
```

```python
test_inputs[numarical_cols]=imputer.transform(test_inputs[numarical_cols])
val_inputs[numarical_cols]=imputer.transform(val_inputs[numarical_cols])
```

```python
test_inputs[numarical_cols].isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(raw_df[numarical_cols])
train_inputs[numarical_cols]=scaler.transform(train_inputs[numarical_cols])
val_inputs[numarical_cols]=scaler.transform(val_inputs[numarical_cols])
test_inputs[numarical_cols]=scaler.transform(test_inputs[numarical_cols])
```

```python
train_inputs[categorical_cols] =train_df[categorical_cols].fillna('unknown')
val_inputs[categorical_cols]=val_df[categorical_cols].fillna('unknown')
test_inputs[categorical_cols] =test_df[categorical_cols].fillna('unknown')
```

```python
train_inputs[categorical_cols].head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
def max_depth_error(md):
  model = DecisionTreeClassifier(random_state=42,max_depth=md)
  model.fit(x_train,train_targets)
  train_error =1 - model.score(x_train,train_targets)
  val_error = 1 - model.score(x_val,val_targets)
  return {'Max Depth': md, 'Train Error': train_error, 'Validation Error': val_error}
```

```python

def predict_input(single_input):
    input_df = pd.DataFrame([single_input])
    input_df[numarical_cols] = imputer.transform(input_df[numarical_cols])
    input_df[numarical_cols] = scaler.transform(input_df[numarical_cols])
    input_df[encoded_col] = encoder.transform(input_df[categorical_cols])
    X_input = input_df[numarical_cols + encoded_col]
    pred = model.predict(X_input)[0]
    prob = model.predict_proba(X_input)[0][list(model.classes_).index(pred)]
    return pred, prob
```
